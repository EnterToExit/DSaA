### Выполнение:

```python
import turtle
import datetime
import math



def create_fractal(func, max_n, doneTurtle = False, step = 200):
    turtle.speed(10000000)
    
    for i in range(1, max_n + 1):
        turtle.reset()
        turtle.speed(10000000)
        start_time = datetime.datetime.now()
        func(step, i)
        total_time = (datetime.datetime.now() - start_time).total_seconds()
        print(f"Глубина: {i} - {round(total_time, 1)}s")
    if doneTurtle: turtle.done()
```

### Кривая Коха


```python
def koch_fractal(step, n):
    if n == 0: 
        turtle.forward(step)
        return
    
    def call_next(): 
        koch_fractal(step / 3, n - 1)
   
    call_next()
    turtle.left(60)
    call_next()
    turtle.right(120)
    call_next()
    turtle.left(60)
    call_next()    

create_fractal(koch_fractal, 5)
```

### Кривая Гильберта


```python
def hilbert_fractal(step, n, dir = 1):
    if n == 0: return
    
    turtle.left(90 * dir)
    hilbert_fractal(step, n - 1, -dir)
    turtle.forward(step)
    
    turtle.right(90 * dir)
    hilbert_fractal(step, n - 1, dir)
    turtle.forward(step)
    
    hilbert_fractal(step, n - 1, dir)
    turtle.right(90 * dir)
    turtle.forward(step)
    
    hilbert_fractal(step, n - 1, -dir)
    turtle.left(90 * dir)

create_fractal(hilbert_fractal, 5, step=20)
```

### Кривая Серпинского


```python
try:
    turtle.reset()
except:
    pass
rules = {"+" : "+", "-" : "-", "X" : "XF-F+F-XF+F+XF-F+F-X", "F" : "F"}


def serp_line_decryption(n, axiom = "F+XF+F+XF"):
    if n == 0:
        return axiom
    else:
        temp = ""
        for i in range(len(axiom)):
            temp += rules[axiom[i]]
        return serp_line_decryption(n - 1, temp)   

    
def draw_serp_line(n, length, t):
    t.clear()
    t.up()                 
    t.seth(0)    
    t.setpos(0,-200)    
    t.down()
    instruction = serp_line_decryption(n)
    for el in instruction:
        if el == "+":
            t.lt(90)
        elif el == "-":
            t.rt(90)
        elif el == "F":
            t.fd(length)
        else:
            pass
    
t = turtle.Turtle()
t.pensize(2) 
t.clear()
t.ht()
turtle.tracer(1,0) 
for i in range(1, 6):
    t1 = time.time()
    draw_serp_line(i,8-i,t)
    t2 = round((time.time() - t1)*1000)
    print(f"Глубина: {i} - {round(t2, 1)}ms ")
turtle.done()
```

### Салфетка Серпинского


```python
def sierpinski_triangle_fractal(step, n, dir = 1):
    
    if n == 0: return
    
    turtle.right(120)
    turtle.forward(step)
    sierpinski_triangle_fractal(step / 2, n - 1, -1)
    turtle.forward(step)
    turtle.right(120 * dir)
    turtle.forward(step)
    
    if dir < 0: sierpinski_triangle_fractal(step / 2, n - 1, -1)   
    turtle.forward(step)
    turtle.right(120 * dir)
    turtle.forward(step)
    
    if dir < 0: sierpinski_triangle_fractal(step / 2, n - 1, -1)
    turtle.forward(step)
    turtle.right(120)
    
create_fractal(sierpinski_triangle_fractal, 5)
```

### Ковер Серпинского


```python
def sierpinski_quad_fractal(step, n):
    
    if n == 0: return
    if n == 1: turtle.begin_fill()
        
    for x in range(4):
        
        for x in range(2):
            sierpinski_quad_fractal(step / 3, n - 1)
            turtle.forward(step)
            
        turtle.forward(step)
        turtle.right(90)
        
    if n == 1: turtle.end_fill()

create_fractal(sierpinski_quad_fractal, 3)
```

### Дерево Пифагора


```python
def pythagoras_tree_iter(step, n, dir = 1):
    
    if n == 0: return
    next_step = math.sqrt((step ** 2) / 2) 
    turtle.forward(step)
    
    
    if n > 1:
        turtle.left(45 * dir)
        pythagoras_tree_iter(next_step, n - 1, dir)
        turtle.left(135 * dir)
        turtle.forward(step)
        turtle.left(45 * dir)
        pythagoras_tree_iter(next_step, n - 1, -dir)
        turtle.right(45 * dir)
        turtle.forward(step)
    else:
        turtle.right(90 * dir)
        turtle.forward(step)
        turtle.right(90 * dir)
        turtle.forward(step)
    turtle.right(90 * dir)
    turtle.forward(step)


def pythagoras_tree(step, n):
    turtle.left(90)
    pythagoras_tree_iter(step, n)


create_fractal(pythagoras_tree, 5)
```
